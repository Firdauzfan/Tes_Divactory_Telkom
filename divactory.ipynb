## 2. BUILD MODEL, MINIMIZE LOG LOSS

### Split Data Cross Validation

Pada bagian ini, kita melatih training set dan menguji model yang sudah dibuat dengan cross validation. <br>
Pembagian data yang dilakukan pada cross validation sebesar `80 %` data training, dan `20 %` data testing. <br>

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

array = df_train.values
X = array[:,0:21]
Y = array[:,21]
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
X_test = df_test

scoring = 'accuracy'
```

### Pemilihan Algoritma

Pada bagian ini, kami menerapkan beberapa algoritma untuk menguji model yang telah dibuat. <br>
Didapatkan hasil akurasi di tiap-tiap algoritma dimana algoritma SVM menghasilkan akurasi yang paling tinggi dibandingkan algoritma yang lain.

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', LinearSVC(C=10)))

results = []
names = []
for name, model in models:
	kfold = model_selection.KFold(n_splits=10, random_state=seed)
	cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
```

### Visualisasi Perbandingan Algoritma

Dilakukan visualisasi terhadap nilai akurasi untuk masing masing algoritma yang digunakan.

```python
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

### Prediksi

Setelah ditentukan algoritma yang terbaik yaitu SVM, dilakukan prediksi langsung untuk mengetahui hasil akurasi dan log loss, ditampilkan juga confussion matrix dan classification report. <br>
Didapatkan **hasil akurasi** : `52 %`, dan **log loss** : `16.501`.

```python
SVM =  LinearSVC(C=10)
SVM.fit(X_train, Y_train)
predictions = SVM.predict(X_validation)
print("Accuracy : {}".format(accuracy_score(Y_validation, predictions)))
print("Log Loss : {}\n".format(log_loss(Y_validation, predictions)))
print("Confusion Matrix : \n{}\n".format(confusion_matrix(Y_validation, predictions)))
print("Classification Report : \n{}".format(classification_report(Y_validation, predictions)))
```
